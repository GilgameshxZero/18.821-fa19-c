```python
import numpy as np
```

```python
P = 7
CMAX = 100
DMAX = 20000
DISPLAY = 35
```

```python
def period(npa):
    for a in range(1, DMAX // 2):
        shift = np.roll(npa, a)
        if np.array_equal(npa[a:], shift[a:]):
            return a
    return None
```

```python
prev = np.array([1 for a in range(DMAX)])

for a in range(CMAX):
    print("Period", period(prev), "for", prev[:DISPLAY])
    prev = np.cumsum(prev) % P
```

```python
def period_v(npa, dmax):
    for a in range(1, dmax // 2):
        shift = np.roll(npa, a)
        if np.array_equal(npa[a:], shift[a:]):
            return a
    return None
```

```python
def calc_v(p, cmax, dmax, display):
    prev = np.array([1 for a in range(dmax)])

    for a in range(cmax):
        print(period_v(prev, dmax), "\t", prev[:display])
        prev = np.cumsum(prev) % p
```

```python
calc_v(5, 100, 20000, 30)
```

```python
calc_v(6, 100, 20000, 30)
```
