```python
def get_patterns(p, n):
    patterns = [[1, p - 1]] if p > 2 else [[1]]
    for _ in range(n - 1):
        pat = [1]
        while True:
            for x in patterns[-1][1:]:
                pat.append((x - pat[-1]) % p)
            if pat[-1] == 0:
                break
            pat.append((1 - pat[-1]) % p)
        
        patterns.append(pat)

    return patterns

get_patterns(p=3, n=9)
```

```python
def count_substrings(patterns, n):
    patterns = ["".join(map(str, pat)) for pat in patterns]
    
    cs = [0]
    for n in range(1, n):
        ss = set()
        for pat in patterns:
            rep = (n // len(pat) + 2) * pat
            for i in range(len(pat)):
                ss.add(rep[i:i + n])
    
        cs.append(len(ss))
    
    return cs

print(count_substrings(get_patterns(p=2, n=100), n=10))
print(count_substrings(get_patterns(p=3, n=100), n=10))
```

```python
for p in [2, 3, 5, 7, 11]:
    cs = count_substrings(get_patterns(p=p, n=1000), n=20)
    diffs = [cs[i] - cs[i - 1] for i in range(2, len(cs))]
    print([diffs[i] - diffs[i - 1] for i in range(1, len(diffs))])
```
