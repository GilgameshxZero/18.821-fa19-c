```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def graph_pascals_mod_p(n, p, **kwargs):
    a = np.zeros(n + 1, dtype=np.int)
    a[0] = 1
    
    for i in range(n):
        y = (n - i) / n
        x_step = 0 if i == 0 else 2 * (1 - y) / i
        x = y
        for j, v in enumerate(a[:i + 1]):
            plt.scatter(x, y, alpha=v/p, c="black",**kwargs)
            
            x += x_step
        
        a[1 : i + 2] += a[:i + 1]
        a %= p
```

```python
%%time
graph_pascals_mod_p(n=100, p=2, s=0.5)
```
