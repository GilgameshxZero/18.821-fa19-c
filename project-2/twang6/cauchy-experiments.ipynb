```python
import numpy as np
import matplotlib.pyplot as plt
```

$$p(x) \sim \frac{1}{x^2 + 1}$$

```python
cauchys = np.random.standard_cauchy(size=10000000)
csum = cauchys.cumsum()
avgs = csum / (np.arange(csum.shape[0], dtype=np.float64) + 1)

plt.hist(cauchys, range=(-25, 25), bins=1024, density=True)
plt.show()
plt.plot(avgs)
plt.show();
```
